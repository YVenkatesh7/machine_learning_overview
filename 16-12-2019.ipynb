```python
from scipy.stats import poisson
```

```python
rv = poisson(2) #avg number of homes - 2
```

```python
rv.pmf(3)  #we want to find out the likelihood that 3 homes will be sold tomorrow
```

```python
import matplotlib.pyplot as plt
```

```python
arr= []
rv = poisson(25)
for num in range(0,40):
    arr.append(rv.pmf(num))
prob = rv.pmf(28)
plt.plot(arr,linewidth = 2.0)
plt.plot([28],[prob],'.',c='r')
plt.show()
```

```python
from scipy.stats import binom
```

```python
n,p = 5,1/6
rv = binom(n,p)
```

```python
rv.pmf(2)
```

```python
fig,ax = plt.subplots(1,1)
x = range(7)
n,p=6,0.5
rv = binom(n,p)
ax.vlines(x,0,rv.pmf(x),colors='k',ls = '--',lw = 2,label ="Probability of Success")
ax.legend(loc = "best")
plt.show()
```
