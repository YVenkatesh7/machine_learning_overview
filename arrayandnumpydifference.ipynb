```python
import time
import sys
import numpy as np
SIZE = 1000
L1 = range(SIZE)
L2 = range(SIZE)
start = time.time() 
result = [(x,y) for x,y in zip(L1,L2)]
print(result)
print((time.time()-start)*1000) 
```
