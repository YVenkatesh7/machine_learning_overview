```python
def f(t):
    return np.exp(-t)*np.sin(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.01)
plt.subplot(221) #middle value rep alignment 1 - vertical 2 - horizontal
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.subplot(222)
plt.plot(t2,np.sin(2*np.pi*t2))
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [1, 4, 9, 6]
labels = ['Frogs', 'Hogs', 'Bogs', 'Slogs']
plt.plot(x,y,'ko--')
# You can specify a rotation for the tick labels in degrees or with keywords.
plt.xticks(x, labels, rotation='vertical')
# Pad margins so that markers don't get clipped by the axes
plt.margins(0.2)
# Tweak spacing to prevent clipping of tick-label
plt.show()
```

```python
#Ticks are the values used to show specific points on the coordinate axis.
#It can be a number or a string.
# values of x and y axes 
x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50] 
y = [1, 4, 3, 2, 7, 6, 9, 8, 10, 5] 
  
plt.plot(x, y, '.') 
plt.xlabel('x') 
plt.ylabel('y') 
  
# 0 is the initial value, 50 is the final value 
# (last value is not taken) and 5 is the difference 
# of values between two consecutive ticks 
plt.xticks(np.arange(0, 51, 5)) 
plt.yticks(np.arange(0, 11, 1)) 
plt.show()
```

```python
#A user defined function is used to format the ticks in millions of dollars on the y axis.
#The default formatter identifies when the x-data being plotted 
#is a small range on top of a large offset
#Tick formatting is controlled by classes derived from Formatter.
#The formatter operates on a single tick value and returns a string to the axis

from matplotlib.ticker import FuncFormatter  #User defined function sets the labels.
x = np.arange(4)
money = [1.5e5, 2.5e6, 5.5e6, 2.0e7]
def millions(x, pos):
    'The two args are the value and tick position'
    return '$%1.2fM' % (x * 1e-6)
formatter = FuncFormatter(millions)
fig, ax = plt.subplots()
ax.yaxis.set_major_formatter(formatter)
plt.bar(x, money)
plt.xticks(x, ('Bill', 'Fred', 'Mary', 'Sue'),rotation = 'vertical')
plt.show()
```
